# Работа с пропусками и выбросами в датасете

Заполнение недостающих данных о клиентах

  *Вспомогательные гистограммы:*


Распределение классов

```python
sns.countplot(data = df, x = 'Class', palette = 'RdPu')
```

Распределение типов путешествия, преобладают бизнес-трипы

```python
sns.countplot(data = df, x = 'Type of Travel', palette = 'RdPu')
```

Вследствие того, что в данных находятся очевидные выбросы (например, возраст - 621 год, задержка прибытия самолета - 15 дней и длина полета 379 тысяч миль/километров), которые могут ухудшить качество модели, очистим датасет от значений, находящихся за трехсигмовыми границами c помощью z-оценки

```python
z = abs(stats.zscore(df[['Age','Flight Distance','Arrival Delay in Minutes', 'Departure Delay in Minutes']], nan_policy='omit'))
df_clean = df[(z<3).all(axis=1)]

df_clean.describe()
```

Пропуски в каждом из признаков заполняются отдельно и по-своему, чтобы не нарушить логику в данных


> **Заполнение пропусков с данными о клиентах**



```python
df_clean['Gender'].dropna()
# отсутствующую информацию о гендере удаляем

df_clean['Age'].fillna(df_clean['Age'].mean().round(0),inplace=True)
# пропуски заполняем средним возрастом

df_clean['Customer Type'].fillna('disloyal Customer',inplace=True)
# считаем, что если нет информации, то клиента нет в базе, и он не является лояльным

df_clean['Type of Travel'].fillna('Business travel',inplace=True)
# заполняем самой популярной категорией, т.к. доля пропусков мала, а доля бизнес-трипов доминирует

df_clean['Class'].dropna()
# строчки с неизвестным классом необходимо удалить

df_clean['Flight Distance'].fillna(df_clean['Flight Distance'].mean().round(0),inplace=True)
df_clean['Departure Delay in Minutes'].fillna(df_clean['Departure Delay in Minutes'].mean().round(0),inplace=True)
df_clean['Arrival Delay in Minutes'].fillna(df_clean['Arrival Delay in Minutes'].mean().round(0),inplace=True)
# Расстояние и время задержки заполняем средними значением
```

Также удалим нулевой возраст клиентов и нулевую длину перелета

```python
df_clean = df_clean[df_clean['Age'] != 0 ]
df_clean = df_clean[df_clean['Flight Distance'] != 0 ]
```



> **Работа с оценками**

Если значение признака больше 5 или данных по нему нет, строчка удаляется. Мы можем совершать такие действия, так как доля таких значений относительно всей генеральной совокупности крайне мала.

```python
df_clean = df_clean[(df_clean.iloc[:,9:23]<=5).all(axis=1)]
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.describe()
```

Теперь все оценки находятся в диапазоне от 0 до 5, число выбросов снижено до минимума и пропусков в датасете нет

```python
sns.countplot(data=df_clean, x='satisfaction', palette = 'RdPu')
```

```python
df_clean['satisfaction'].value_counts()
```

Доля низвестной категории '-' составлет примерно 20%, что является достаточно значимым объемом информации для обучения модели, несмотря на итак большой размер датасета.
Посмотрим среднее значение всех оценок каждого клиента в трех категориях удовлетворенности:

```python
mean__ = df_clean[df_clean['satisfaction']=='-'].iloc[:,9:23].mean(axis = 1)
mean_s = df_clean[df_clean['satisfaction']=='satisfied'].iloc[:,9:23].mean(axis = 1)
mean_n = df_clean[df_clean['satisfaction']=='neutral or dissatisfied'].iloc[:,9:23].mean(axis = 1)

[mean__.mean(),mean_s.mean(),mean_n.mean()]
```

```python
a = mean__
b = mean_s
c = mean_n

plt.hist(b,alpha=0.9, label='Удовлетворены',color = 'plum', bins =16)
plt.hist(c,alpha=0.5, label='Не удовлетворены',color = 'purple', bins =16)
plt.hist(a, alpha=0.5, label='Неизвестная категория',color = 'yellow', bins =16)

plt.legend(loc='upper left')
plt.show()
```

Видно, что нельзя "присоединить" неизвестную категорию к двум другим, так как, исходя из среднего значения, эта группа находится между ними. Поэтому для улучшения качества потенциальной модели, удалим третью категорию

```python
df_final = df_clean[df_clean['satisfaction'] !='-']
```

Корреляция между признаками

```python
dfc = df_final.drop(['Gender','Customer Type','Type of Travel','Class','satisfaction'], axis = 1)
corr = dfc.corr().round(1)
corr = corr[(corr >=.4) | (corr <=-.4)]
sns.heatmap(corr, annot = True, cmap = 'PuRd', vmin = -1)
```

Данные межу собой коррелируют незначительно, а там, где линейная связь выше 0,4, скорее всего, обусловлено случайности. Например, вряд ли имеется причинно-следственная связь между чистотой салона и развлечением в самолете

# Перевод категориальных признаков в бинарные

Посмотрим, какие признаки нуждаются в дальнейшей обработке

```python
df_final.describe(include =[object])
```

```python
df_final.tail(2)
```

Гендер, лояльность, тип перелета и удовлетворенность уже находятся в бинарном виде, необходимо лишь заменить значения на 1 и 0.

Класс пассажира нужно перевести в бинарный вид с помощью One-Hot Encoding

```python
df_final['Gender'] = df_final['Gender'].map({'Male' : 1, 'Female' : 0})
df_final['Customer Type'] = df_final['Customer Type'].map({'Loyal Customer' : 1, 'disloyal Customer' : 0})
df_final['Type of Travel'] = df_final['Type of Travel'].map({'Business travel' : 1, 'Personal Travel' : 0})
df_final['satisfaction'] = df_final['satisfaction'].map({'satisfied' : 1, 'neutral or dissatisfied' : 0})

df_final.head(2)
```

Также проведем масштабирование данных



```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler

categorical = ['Class']
numeric_features = [col for col in df_final.columns if col not in categorical]

column_transformer = ColumnTransformer([
    ('ohe', OneHotEncoder(drop='first', handle_unknown="ignore"), categorical),
    ('scaling', MinMaxScaler(), numeric_features)
])

df_final_trans= column_transformer.fit_transform(df_final)

lst = list(column_transformer.transformers_[0][1].get_feature_names_out())
lst.extend(numeric_features)

df_final_trans = pd.DataFrame(df_final_trans, columns=lst)

df_final_trans.head()
```

# Обучение модели

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
```

```python
X = df_final_trans.drop(['id', 'satisfaction'], axis=1, inplace=True)
Y = df_final['satisfaction'] # таргет
```

```python
X = df_final_trans
X.head(3)
```

Разобъем генеральную совокупность на две выборки в соотношении 75% на 25%

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=16)

X_train.shape, X_test.shape
```

Обучение модели

```python
model = LogisticRegression()

model.fit(X_train, y_train)

pred = model.predict(X_test)
```

```python
pred[:5]
```

Проверяем модель на адекватность

```python
confusion_matrix(y_test, pred)
```

**Accuracy**

```python
accuracy_score(y_test, pred)
```

**Recall**

```python
recall_score(y_test, pred)
```

**Precision**

```python
precision_score(y_test, pred)
```

Посмотрим на изменение метрик при изменении порога
СТоиит отметить, что в данных нет четкого дисбаланса в таргете, то есть метрика Accuracy может отражать качество модели

```python
probs = model.predict_proba(X_test)

classes = probs[:,1] > 0.3

res = pd.DataFrame({'Recall': [recall_score(y_test, classes)], 'Accuracy': [accuracy_score(y_test, classes)], 'Precision': [precision_score(y_test, classes)]}).T
res, confusion_matrix(y_test, classes)
```
